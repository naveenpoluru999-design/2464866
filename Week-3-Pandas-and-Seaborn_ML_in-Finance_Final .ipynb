```python
import pandas as pd
import matplotlib.pyplot as plt


sid = 2464866  # change this to your SID


df = pd.DataFrame({
    f'Col{i}': range(1, 11) for i in range(10)
})



sid_str = str(sid)
col1 = int(sid_str[-2])  # second-to-last digit
col2 = int(sid_str[-1])  # last digit


if col1 == col2:
    col2 = (col2 + 1) % 10

print(f"Using columns: {col1} and {col2}")

# ---- Step 4: Select columns ----
x = df.iloc[:, col1]
y = df.iloc[:, col2]


plt.figure(figsize=(6, 6))
plt.scatter(x, y, c='blue', label=f'Column {col1} vs {col2}')
plt.xlabel(f'Column {col1}')
plt.ylabel(f'Column {col2}')
plt.title(f'Bicolour Diagram (SID: {sid})')
plt.legend()
plt.grid(True)
plt.show()
```
